## Seleções e Frequências

### Selecionando somente os imóveis classificados como tipo "Apartamento"

```python
import pandas as pd
```

```python
# Importando
dados = pd.read_csv('aluguel.csv', sep=';')
dados
```

```python
apartamentos = dados[dados['Tipo'] == 'Apartamento']
apartamentos
```

```python
#apartamentos.to_csv('Apartamentos.csv', sep=';', index=False)
```

### Selecionando os imóveis classificados com tipos 'Casa', 'Casa de Condomínio', e 'Casa de Vila'

```python
casas = dados[(dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')] 
casas
```

```python
casas.index = range(casas.shape[0])
casas
```

```python
#casas.to_csv('casas.csv', sep=';', index=False)
```

### Selecionando os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites

```python
dados
```

```python
dados_area_entre_60_e_100 = dados[(dados['Area'] >= 60) & (dados['Area'] <= 100)]
dados_area_entre_60_e_100
```

```python
#dados_area_entre_60_e_100.to_csv('Dados_Com_Area_Entre_60_E_100.csv', sep=';', index=False)
```

### Selecionando os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00

```python
dados
```

```python
acima_de_4_quartos_e_aluguel_menor_que_2000 = dados[(dados['Quartos'] >= 4) & (dados['Valor'] < 2000)]
acima_de_4_quartos_e_aluguel_menor_que_2000
```

```python
acima_de_4_quartos_e_aluguel_menor_que_2000.shape[0]
```

```python
print("Nº de imóveis classificados com tipo 'Apartamento' -> {}".format(apartamentos.shape[0]))
print("Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila' -> {}".format(casas.shape[0]))
print("Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {}".format(dados_area_entre_60_e_100.shape[0]))
print("Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {}".format(acima_de_4_quartos_e_aluguel_menor_que_2000.shape[0]))
```
